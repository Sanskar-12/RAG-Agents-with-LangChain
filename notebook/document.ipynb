```python
### Create  a simple txt file

import os

os.makedirs("./data/text_files",exist_ok=True)
```

```python
# Meaningful text content
text = """Agentic AI systems are designed to perform tasks autonomously by combining reasoning, memory, and tool usage.

In modern applications, agents can translate natural language into structured queries such as SQL, validate them, and retrieve real-time data from databases.

This approach enables dynamic data retrieval without relying on traditional embedding-based retrieval methods, making it efficient for structured data systems.

Such systems are widely used in applications like chatbots, analytics platforms, and intelligent assistants."""

# File path
file_path = "./data/text_files/sample.txt"

# Write to file
with open(file_path, "w") as f:
    f.write(text)

print("File created at:", file_path)
```

```python
### Text Loader to read the text and convert into Document

from langchain_community.document_loaders import TextLoader

loader=TextLoader("./data/text_files/sample.txt",encoding="utf-8")

document=loader.load()

print(document)
```
